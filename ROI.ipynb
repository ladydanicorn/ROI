```python
class Income:
    def __init__(self, rental, laundry=0, storage=0, misc=0):
        self.rental = rental
        self.laundry = laundry
        self.storage = storage
        self.misc = misc

    def total_income(self):
        return self.rental + self.laundry + self.storage + self.misc


class Expenses:
    def __init__(self, taxes, insurance, utilities, mortgage, vacancy=0, repairs=0, capex=0, management=0, hoa=0):
        self.taxes = taxes
        self.insurance = insurance
        self.utilities = utilities
        self.mortgage = mortgage
        self.vacancy = vacancy
        self.repairs = repairs
        self.capex = capex
        self.management = management
        self.hoa = hoa

    def total_expenses(self):
        return (self.taxes + self.insurance + self.utilities + self.mortgage +
                self.vacancy + self.repairs + self.capex + self.management + self.hoa)


class Investment:
    def __init__(self, down_payment, closing_costs, rehab_budget, misc_other=0):
        self.down_payment = down_payment
        self.closing_costs = closing_costs
        self.rehab_budget = rehab_budget
        self.misc_other = misc_other

    def total_investment(self):
        return self.down_payment + self.closing_costs + self.rehab_budget + self.misc_other


class Property:
    def __init__(self, income, expenses, investment):
        self.income = income
        self.expenses = expenses
        self.investment = investment

    def monthly_cashflow(self):
        return self.income.total_income() - self.expenses.total_expenses()

    def annual_cashflow(self):
        return self.monthly_cashflow() * 12

    def roi(self):
        return (self.annual_cashflow() / self.investment.total_investment()) * 100

# Collect data from user
rental_income = float(input("Enter rental income: "))
laundry_income = float(input("Enter laundry income: "))
storage_income = float(input("Enter storage income: "))
misc_income = float(input("Enter miscellaneous income: "))

taxes = float(input("Enter taxes: "))
insurance = float(input("Enter insurance: "))
utilities = float(input("Enter utilities: "))
mortgage = float(input("Enter mortgage payment: "))
vacancy = float(input("Enter vacancy costs: "))
repairs = float(input("Enter repair costs: "))
capex = float(input("Enter CapEx (capital expenditures): "))
management = float(input("Enter property management fees: "))
hoa = float(input("Enter HOA fees: "))

down_payment = float(input("Enter down payment: "))
closing_costs = float(input("Enter closing costs: "))
rehab_budget = float(input("Enter rehab budget: "))
misc_other = float(input("Enter miscellaneous other costs: "))

# Create instances based on input data
income = Income(rental=rental_income, laundry=laundry_income, storage=storage_income, misc=misc_income)
expenses = Expenses(taxes=taxes, insurance=insurance, utilities=utilities, mortgage=mortgage, vacancy=vacancy, repairs=repairs, capex=capex, management=management, hoa=hoa)
investment = Investment(down_payment=down_payment, closing_costs=closing_costs, rehab_budget=rehab_budget, misc_other=misc_other)

# Create Property instance and calculate ROI
property = Property(income, expenses, investment)

print(f"Monthly Cashflow: ${property.monthly_cashflow()}")
print(f"Annual Cashflow: ${property.annual_cashflow()}")
print(f"ROI: {property.roi()}%")
```
